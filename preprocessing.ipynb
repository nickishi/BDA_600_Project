```python
import pandas as pd
import re
import emoji
import numpy as np
```

```python
#we will read in our file that merges all files
#from extract.ipynb 
df = pd.read_csv('dataset.csv')
df
```

```python
#Create new dataframe to compare changes
tw_list = pd.DataFrame(df)
tw_list["cleaned"] = tw_list["text"].apply(str)
tw_list
```

```python
clean_tweets = []
for tweet in tw_list["cleaned"]:
    tweet = re.sub('@[^\s]+', '',tweet) #Remove @ sign
    tweet = re.sub('_[A-Za-z0-9]+', '',tweet) #Remove usernames with additional _underscore
    tweet = tweet.replace('#', '') #Remove hashtag (#) but keep the text
    tweet = re.sub(r'(?:\@|http?\://|https?\://|www)\S+', '', tweet) #Remove http links
    tweet = ''.join(c for c in tweet if c not in emoji.EMOJI_DATA) #Remove Emojis
    tweet = re.sub('RT', '',tweet) #Remove tweets that have 'RT'
    tweet = re.sub(r'[^\w\s]', '', tweet) #remove special characters
    tweet = tweet.replace('\n', '') #remove line breaks
    clean_tweets.append(tweet)
tw_list["cleaned"] = clean_tweets
```

```python
tw_list
```

```python
tw_list.drop_duplicates(subset=['cleaned'])
```

```python
#add a tweet_id column in index 0
tw_list['tweet_id'] = np.zeros(len(tw_list))

#autoincrement tweet ID starting from 1
tw_list['tweet_id'] = range(1, len(tw_list)+1)

# move 'tweet_id' column to index 0
cols = tw_list.columns.tolist()
cols = cols[-1:] + cols[:-1]
tw_list = tw_list[cols]

#print dataframe to verify
tw_list.head()
```

```python
#remove text column
tw_list = tw_list.drop('text', axis=1)

#rename the cleaned column to 'text'
tw_list = tw_list.rename(columns={'cleaned': 'text'})

#print dataframe to verify
tw_list.head()
```

```python
#save to csv
tw_list.to_csv('data_clean.csv')
```
